```
import xgboost as xgb
import pandas as pd
```

```
df = pd.DataFrame({'x': [1, 2, 10], 'y': [1,1,0]})
dmat = xgb.DMatrix(df[['x']], df['y'])
model = xgb.train({'eta': 1, 'reg_lambda': 0}, dmat, num_boost_round=1)
print(model.get_dump()[0])
```

```
df = pd.DataFrame({'x': [1,1.1,1.2, 2.2,2.3,2.3, 10.5,10.5,10.6], 'y': [1,1,1, 0,1,1, 0,0,0]})
dmat = xgb.DMatrix(df[['x']], df['y'])
model = xgb.train({'eta': 1, 'reg_lambda': 0}, dmat, num_boost_round=1)
print(model.get_dump()[0])
```

```
from sklearn.datasets import load_breast_cancer
X, y = load_breast_cancer(return_X_y=True)
dmat2 = xgb.DMatrix(X, y)
model2 = xgb.train({'eta': 1, 'reg_lambda': 0}, dmat2, num_boost_round=1)
print(model2.get_dump()[0][:100], '...')
```
