```
import xgboost as xgb
import pandas as pd
import numpy as np
from sklearn.datasets import load_wine
```

```
wine = load_wine()
```

```
y = wine['target']==0
```

```
X = wine['data']
```

```
# No regularization (gets to a full tree):
model = xgb.train(params={'n_estimators': 1,
                          'max_depth': 3,
                          'reg_lambda': 0},
                  num_boost_round=1,
                  dtrain=xgb.DMatrix(X,y))
model.trees_to_dataframe()
```

```
# (huge) L1 penalty
model = xgb.train(params={'n_estimators': 1,
                          'max_depth': 3,
                          'reg_alpha': 1000},
                  num_boost_round=1,
                  dtrain=xgb.DMatrix(X,y))
model.trees_to_dataframe()
```

```
# (huge) L2 penalty
model = xgb.train(params={'n_estimators': 1,
                          'max_depth': 3,
                          'reg_lambda': 1000},
                  num_boost_round=1,
                  dtrain=xgb.DMatrix(X,y))
model.trees_to_dataframe()
```

```
# (huge) leaf penalty
model = xgb.train(params={'n_estimators': 1,
                          'max_depth': 3,
                          'gamma': 1000},
                  num_boost_round=1,
                  dtrain=xgb.DMatrix(X,y))
model.trees_to_dataframe()
```

```
# default penalties (gets nearly to a full tree; just Leaf 3 doesn't split)
model = xgb.train(params={'n_estimators': 1,
                          'max_depth': 3},
                  num_boost_round=1,
                  dtrain=xgb.DMatrix(X,y))
model.trees_to_dataframe()
```
